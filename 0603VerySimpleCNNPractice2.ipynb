```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.autograd import Variable
```

```python
input = torch.ones(1,1,3,3)
input = Variable(input,requires_grad=True)
```

```python
func= nn.Conv2d(1,1,3)
```

```python
func.weight
```

```python
out = func(input)
```

```python
print(out)
```

```python
print(input.grad)
print(input.grad_fn)
```

```python
out.backward()
```

```python
print(input.grad)
print(input.grad_fn)
```
