```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.autograd import Variable
```

```python
input = torch.ones(1,1,3,3)
filter = torch.ones(1,1,3,3)
```

```python
input = Variable(input, requires_grad=True)
print(input)
```

```python
filter = Variable(filter)
out = F.conv2d(input,filter)
```

```python
print(out)
```

```python
out.backward()
```

```python
print(out.grad_fn)
```

```python
print(input.grad)
```

```python
input2 = Variable(torch.ones(1,1,3,3,),requires_grad=True)
```

```python
fileter2 = filter +1
```

```python
out = F.conv2d(input2,fileter2)
print(out)
```

```python
out.backward()
print(out.grad_fn)
print(out.data)
print(out.grad)
```
