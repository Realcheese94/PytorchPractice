```python
import torch
from torch.autograd import Variable
```

```python
a = torch.ones(2,2)
print(a)
```

```python
a = Variable(a,requires_grad=True)
print(a)

```

```python
print(a.data)
print(a.grad)
print(a.grad_fn)
```

```python
b = a +2
print(b)
```

```python
c = b**2
print(c)
```

```python
out= c.sum()
print(out)
out.backward()
```
