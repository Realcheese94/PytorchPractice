```python
import torch
```

```python
x = torch.rand(4,3)
y = torch.rand(4,3)
result = torch.cat([x,y],dim=0)
```

```python
#x와 y 그리고 두개를 합친 result 를 출력
print(x)
print(y)
print(result)
```

```python
#result1, result2 는 result함수를 dim 이 1 인값으로 자르는 것이다.
result1,result2 = torch.chunk(result,2,dim=1)
result1
result2
```

```python
#result3 은 result를 자른 것이다.
result3 = torch.split(result,1,dim=0)
result3
```

```python
#torch.squeeze 와 torch.unsqueeze 연습
x = torch.rand(3,1,2,5)
print("x는 ,,,,,,,,,,,,,,,,,,,,,,,,,,,,,,,,,,",x)
sq_x = torch.squeeze(x)
print("sq_x................................",sq_x)
usq_x = torch.unsqueeze(sq_x,dim=0)
print("usq_x............................",usq_x)
```
